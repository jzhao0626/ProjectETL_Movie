```python
import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
emmy_df= pd.read_csv('emmys.csv').dropna()
stream_df = pd.read_csv('streaming platforms.csv').dropna()
```

```python
stream_clean = stream_df.rename(columns = {'Title' : 'nominee'})
stream_clean['nominee'] = stream_clean['nominee'].str.lower()
emmy_df['nominee'] = emmy_df['nominee'].str.lower()
```

```python
merged = pd.merge(stream_clean, emmy_df, on = "nominee")
```

```python
merged_filtered= merged.loc[merged['Year'] >= 2013]
```
